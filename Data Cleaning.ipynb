```python
'''We can handle negative values by ignoring them, since they are few in number. After filtering them out, 
   we will store the cleaned data back into the same DataFrame'''
df_bookings = df_bookings[df_bookings.no_guests>0]
df_bookings
```

```python
#Find the min & max of revenue generated
print(f"Minimum_Bookings:- {df_bookings.revenue_generated.min()}"),
print(f"Maximum_Bookings:- {df_bookings.revenue_generated.max()}"),
print(f"Average_Bookings:- {df_bookings.revenue_generated.mean()}")
#These nos are incorrect beacuse in bookings dataframes it refers to single booking id & how can in single booking id it can get this much of revenue
```

```python
#To handle this error we have to do 3 std deviation outlier
#It is 1 standard deviation
avg,std = df_bookings.revenue_generated.mean(),df_bookings.revenue_generated.std()
#Outlier => It is data point that is too extreme to remove outlier & we use techniques called standard deviation
```

```python
avg,std
```

```python
#If any value is more than 3 standard deviation away from mean(which is an average) then we can filter it out
#To get 3 standard deviation
higher_limit = avg+ 3*std
higher_limit
```

```python
lower_limit = avg - 3*std
lower_limit
```

```python
#To check whether there is any negative revenue
df_bookings[df_bookings.revenue_generated<0]
```

```python
#To check the higher limit
df_bookings[df_bookings.revenue_generated>higher_limit]
```

```python
#To handle this error we can ignore it because it doesnot affect anything in eda as these values are few & to do this 
df_bookings = df_bookings[df_bookings.revenue_generated<higher_limit]
df_bookings
#after the cleaning store it to original data frame
```
